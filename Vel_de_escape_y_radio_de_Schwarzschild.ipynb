```python
def vel_esc(M,R):
  return np.sqrt((2.0*G*M)/(R))
```

```python
c = 3.0E8 #m/s
G = 6.67E-11 #Nm^2/kg^2

M_T = 5.97E24 #kg
R_T = 6371 #km (promedio)
R_T_m = R_T*1000 #m (promedio)

vel_esc_T = vel_esc(M_T,R_T_m) #m/s
vel_T_km_per_s = vel_esc_T/1000 #km/s
print(f"La velocidad de escape de la Tierra es  {vel_T_km_per_s:.2f} km/s")
```

```python
M_sun = 1.99E30 #kg
R_sun = 696340 #km (promedio)
R_sun_m = R_sun*1000 #m (promedio)

vel_esc_sun = vel_esc(M_sun,R_sun_m) #m/s
vel_sun_km_per_s = vel_esc_sun/1000 #km/s
print(f"La velocidad de escape del Sol es {vel_sun_km_per_s:.2f} km/s")
```

# **Velocidad de escape**

Para una estrella con el radio de nuestro Sol (R$_{\odot}=$ 696340 km), la velocidad de escape crece con la masa de esta forma:

```python
array = np.logspace(-5, 6, 100)
M_array = M_sun*array
#print(M_array)

import matplotlib.pyplot as plt

fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(1, 1, 1)

g_xlim = [1E-5, 1E6]
g_ylim = [1,600000]
g_xlabel = "M (M$_{\odot}$)"
g_ylabel = "v$_{esc}$ (km/s)"

ax.set_xlim(g_xlim)
ax.set_ylim(g_ylim)
plt.hlines(300000, 1E-5, 1E6, colors='black', linestyles = 'dashed')
ax.set_xscale('log')
#ax.set_yscale('log')
ax.set_xlabel(g_xlabel, fontsize=15)
ax.set_ylabel(g_ylabel, fontsize=15)
plt.plot(array,0.001*vel_esc(M_array,R_sun_m),lw=3)
plt.savefig('test_radio_fijo.png')
```

Ahora, si mantenemos la masa del Sol (M$_{\odot}$ = 1.99$\times$10$^{30}$ kg), pero empezamos a comprimirlo manteniendo su forma esfÃ©rica, el resultado serÃ­a este:

```python
ar_2 =  np.logspace(-3, 0, 100)
R_array = R_sun*ar_2

import matplotlib.pyplot as plt

fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(1, 1, 1)

g_xlim = [1E-3, 1.0]
g_ylim = [1,600000]
g_xlabel = "R (R$_{\odot}$)"
g_ylabel = "v$_{esc}$ (km/s)"

ax.set_xlim(g_xlim)
ax.set_ylim(g_ylim)
plt.hlines(300000, 1E-3, 1.0, colors='black', linestyles = 'dashed')
#plt.ylim()
ax.set_xscale('log')
#ax.set_yscale('log')
ax.set_xlabel(g_xlabel, fontsize=15)
ax.set_ylabel(g_ylabel, fontsize=15)
plt.plot(ar_2,0.001*vel_esc(M_sun,R_array),color='red')
plt.savefig('test_masa_fija.png')
```

# **Radio de Schwarzschild**

```python
def schw_rad(M):
  return (2.0*G*M)/(c**2)

# Calculado en m
```

```python
# Tu masa. Por ejemplo,
M_hum = 60 #kg

print(f"Tu radio de Schwarzschild es " + str(1000*schw_rad(M_hum)) + " mm")
print(f"El radio de Schwarzschild de la Tierra es " + str(1000*schw_rad(M_T)) + " mm")
print(f"El radio de Schwarzschild del Sol es " + str(0.001*schw_rad(M_sun)) + " km")
```
